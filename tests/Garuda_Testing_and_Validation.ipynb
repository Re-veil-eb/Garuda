## Inspect submission

```
import pandas as pd

df = pd.read_csv("sample_submission.csv")

print("Rows:", len(df))
display(df.head(10))
display(df.tail(10))
```

## Verify ranking constraints

```
assert len(df) == 100
assert df["candidate_id"].is_unique
assert df["rank"].is_unique
assert list(df["rank"]) == list(range(1,101))
assert all(df["score"].iloc[i] >= df["score"].iloc[i+1] for i in range(len(df)-1))

print("✅ Ranking checks passed")
```
