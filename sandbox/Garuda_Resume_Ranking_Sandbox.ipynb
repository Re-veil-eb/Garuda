Once your Google Drive is mounted, you can interact with it as a regular file system. Let's create a directory in your Google Drive to store the submission files and then copy them.

````python
import os

# Define the path to your Google Drive
drive_path = '/content/drive/MyDrive'

# Define the folder name for your submission in Google Drive
submission_folder = os.path.join(drive_path, 'Garuda_submission')

# Create the folder if it doesn't exist
os.makedirs(submission_folder, exist_ok=True)

# List of files to copy from the current directory to Google Drive
files_to_copy = [
    'submission.csv',
    'sample_submission.csv',
    'edge_case_submission.csv'
]

# Copy each file
for file_name in files_to_copy:
    source_path = file_name
    destination_path = os.path.join(submission_folder, file_name)
    try:
        os.system(f'cp {source_path} {destination_path}')
        print(f'Copied {file_name} to {submission_folder}')
    except Exception as e:
        print(f'Error copying {file_name}: {e}')

print(f"All specified files have been copied to '{submission_folder}' in your Google Drive.")
````

````python
import os
import sys
import platform
import subprocess
import psutil

# Detect CPU cores
cpu_cores = os.cpu_count()

# Detect RAM in GB
try:
    ram_gb = round(psutil.virtual_memory().total / (1024**3))
except Exception:
    ram_gb = 'Unknown (psutil might not be installed or accessible)'

# Detect Python version
python_version = sys.version.split()[0]

# Detect OS
os_info = platform.platform()

# Detect GPU for inference
uses_gpu_for_inference = False
try:
    # Check if nvidia-smi command exists and runs
    result = subprocess.run(['nvidia-smi'], capture_output=True, text=True, check=True)
    if "NVIDIA-SMI" in result.stdout:
        uses_gpu_for_inference = True
except (subprocess.CalledProcessError, FileNotFoundError):
    uses_gpu_for_inference = False

print(f"Detected CPU Cores: {cpu_cores}")
print(f"Detected RAM: {ram_gb} GB")
print(f"Detected Python Version: {python_version}")
print(f"Detected OS: {os_info}")
print(f"Detected GPU: {uses_gpu_for_inference}")
````
